```python
# reading the input file
filepath=r'C:\Users\Ram Kumar R S\Downloads\homo_variants\CYCS_dog.txt'
def read_file(filepath): # function definition
    d = dict() # dict to store the fasta file

    with open(filepath, 'r') as file:
        lines = file.readlines()
        for line in lines:    # fasta file reading process
            line = line.strip()
            if line.startswith('>'):
                key = line
                d[key] = ''
            else:
                d[key] += line
    return d

dog_seq=read_file(filepath) # function calling
#dog_exon_seq={k:v for k,v in dog_seq.items() if "TGE" in k}
dog_exon_seq=''.join([v for k,v in dog_seq.items() if "TGE" in k]) # filtering only for "TGE" in headers

def mRNA(line_file):
    mRNAs=line_file.replace('T',"U")  # function for transcription process
    return mRNAs
mRNAseq=mRNA(dog_exon_seq) # function calling
mRNAseq=mRNAseq[::-1] # sequence reversing operation
codon=[mRNAseq[i:i+3] for i in range(0,len(mRNAseq),3)]
#protein=''.join([dic_protein[i] for i in codon])
protein=''.join([dic_protein.get(i,'n') for i in codon]) # list comprehension method of protein translation for codons
                                                         # present in the container "codon" else default value "n" returns
print('The protein sequence generated from Dog Cytochrome gene is as follows:\n{}.'.format(protein))
```


# Permutation performed for cytochrome dog file

```python
import itertools
data_dog_exons={}  # Dict to store the exon seq of the dog file
with open (r'C:\Users\Ram Kumar R S\Downloads\CYCS_dog_exons.txt','r') as file:
    lines = file.readlines()
    for line in lines:
            line = line.strip()   
            if line.startswith('>'):
                key = line
                data_dog_exons[key] = ''
            else:
                data_dog_exons[key] += line
    

dog_exon_seq={k:v for k,v in data_dog_exons.items() if "exon" in k}  # filtering process to extract "exon" alone
exon_seq=([i for i in dog_exon_seq.values()]) # exon sequence 
exon_header=[i for i in dog_exon_seq.keys()]  # exon header

header_combo=[list(x) for i in range(1,4) for x in itertools.permutations(exon_header,i)]
sequence_combo=[list(x) for i in range(1,4) for x in itertools.permutations(exon_seq,i)]

header_combo_dic=[(i,x) for i in range(1,4) for x in itertools.permutations(exon_header,i)] # creates permutation 
sequence_combo_dic=[(i,x) for i in range(1,4) for x in itertools.permutations(exon_seq,i)]  # writh track of what no. 
                                                                                            # times getting permuted 'i'

# File writing process
counter=0 # count variable to find the no. of items
with open(r'C:\Users\Ram Kumar R S\Downloads\homo_variants\Permutations_output.txt','w') as fout:
    for i in range(15):
        counter+=1
        fout.write("This is permuting {} element(s).\n".format(header_combo_dic[i][0]))
        fout.write('{}\n'.format(','.join(header_combo_dic[i][-1])))
        fout.write('{}\n'.format(','.join(sequence_combo_dic[i][-1])))
        fout.write('*'*100)
        fout.write('\n')
    fout.write('There are {} number of combinations in this file.'.format(counter))
print('Job successful!')
```
